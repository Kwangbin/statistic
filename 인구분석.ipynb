```python
import pandas as pd
#폰트 지정
import seaborn as sns
sns.set(font="NanumBarunGothic")
```

```python
url ="https://ko.wikipedia.org/wiki/%EB%8C%80%ED%95%9C%EB%AF%BC%EA%B5%AD%EC%9D%98_%EC%9D%B8%EA%B5%AC"
url
```

```python
table=pd.read_html(url)
```

```python
df = table[4]
```

```python
sns.pointplot(data=df , x="연도 (년)",y="추계인구(명)")
```

```python
#그래프 사이즈 조정
# matplotlib 임포트
import matplotlib.pyplot as plt
 
```

```python
# 그래프 사이즈 조정
# 뒤에 그림판을 지정 하는 원리

plt.figure(figsize= (15,4))
sns.pointplot(data=df , x="연도 (년)",y="추계인구(명)")
```

```python
# x축 글씨가 겹치니 각도를 비스듬이 표시


plt.figure(figsize= (15,4))
plt.xticks(rotation=60)
sns.pointplot(data=df , x="연도 (년)",y="추계인구(명)")
```

```python
# x축 글씨가 겹치니 각도를 비스듬이 표시


plt.figure(figsize= (15,4))
plt.xticks(rotation=90)
sns.pointplot(data=df , x="연도 (년)",y="추계인구(명)")
```

```python
#lineplot

plt.figure(figsize= (15,4))
plt.xticks(rotation=60)
sns.lineplot(data=df , x="연도 (년)",y="추계인구(명)")
```

```python
# 사망자수
plt.figure(figsize= (15,4))
plt.xticks(rotation=90)
sns.pointplot(data=df , x="연도 (년)",y="사망자수(명)")
```

```python
# 출생 자수와 사망 자수를 한화면에 표시

plt.figure(figsize= (15,4))
plt.xticks(rotation=60)
sns.pointplot(data=df , x="연도 (년)",y="출생자수(명)")
sns.pointplot(data=df , x="연도 (년)",y="사망자수(명)")
```

```python
# 출생 자수와 사망 자수를 한화면에 표시
# 사망자수의 선색 변경

plt.figure(figsize= (15,4))
plt.xticks(rotation=60)
sns.pointplot(data=df , x="연도 (년)",y="출생자수(명)")
sns.pointplot(data=df , x="연도 (년)",y="사망자수(명)", color="orange")
```

```python
# 출생 자수와 사망 자수를 한화면에 표시
# 화면 y축 4에서 8로 면경

plt.figure(figsize= (15,8))
plt.xticks(rotation=60)
sns.pointplot(data=df , x="연도 (년)",y="출생자수(명)")
sns.pointplot(data=df , x="연도 (년)",y="사망자수(명)", color="orange")
```
